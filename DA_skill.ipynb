# Flatten the skills into a single list

```python
from ast import literal_eval 
from collections import Counter
```

```python
# Convert the string representations of lists to actual lists
df['description_tokens'] = df['description_tokens'].apply(literal_eval)
```

```python
# Flatten the lists of skills into a single list
all_skills = [skill for sublist in df['description_tokens'] for skill in sublist]
```

```python
# Count the occurrences of each skill
skill_counts = Counter(all_skills)
```

```python
# Create a new DataFrame with the skill names and their counts
skills_df = pd.DataFrame(skill_counts.items(), columns=['skill', 'counts'])
skills_df['Percentage'] = (skills_df['counts'] / len(df)) * 100
skills_df = skills_df.sort_values(by='Percentage', ascending=False)
```

# Top 5 Skills for Data Analysts

```python
top5_df=skills_df.nlargest(5, 'counts')
top5_df
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize = (5,5))
sns.barplot(data = top5_df, x = 'skill', y = 'counts')
plt.title('Top 5 Skills for Data Analyst')
plt.xlabel('Skills')
plt.ylabel('Counts')
plt.show()
```
